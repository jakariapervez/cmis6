```python
def consolidateValue(df,col_index,imed_code):   
    mycat=df.loc[df['Imed_Code']==imed_code]
    #mycat=mycat.iloc[:,col_index]
    #print(mycat)
    Total=0
    for rowindex,row in mycat.iterrows():
        Total += row[col_index]
    return Total
         
        
   
        
        
    
    
```

```python
def createCumulativeMonthlyExpenditure(df):
    cum_expenditure=createCumulativeDaraframe()
    cum_expenditure['Jul']=df['Jul']
    cum_expenditure['Aug']= cum_expenditure['Jul']+df['Aug']
    cum_expenditure['Sep']= cum_expenditure['Aug']+df['Sep']
    cum_expenditure['Oct']= cum_expenditure['Sep']+df['Oct']
    cum_expenditure['Nov']= cum_expenditure['Oct']+df['Nov']
    cum_expenditure['Dec']= cum_expenditure['Nov']+df['Dec']
    cum_expenditure['Jan']= cum_expenditure['Dec']+df['Jan']
    cum_expenditure['Feb']= cum_expenditure['Jan']+df['Feb']
    cum_expenditure['Mar']= cum_expenditure['Feb']+df['Mar']
    cum_expenditure['Apr']= cum_expenditure['Mar']+df['Apr']
    cum_expenditure['May']= cum_expenditure['Apr']+df['May']
    cum_expenditure['Jun']= cum_expenditure['May']+df['Jun']
    cum_expenditure['Q1']=cum_expenditure['Sep']
    cum_expenditure['Q2']=cum_expenditure['Dec']
    cum_expenditure['Q3']=cum_expenditure['Mar']
    cum_expenditure['Q4']=cum_expenditure['Jun']    
    return cum_expenditure
    
    
    
   
    
```

```python
def createPrevData(dpp_alloc,weight,finan_prev,phy_prev,finan_curr,phy_curr):
    df=pd.DataFrame()
    df=df.assign(Dpp_alloc=dpp_alloc)
    df=df.assign(Weight=weight)
    df=df.assign(Fin_prev=finan_prev)
    df=df.assign(Phy_prev=phy_prev)
    df=df.assign(Fin_curr=finan_curr)
    df=df.assign(Phy_curr=phy_curr)
    return df
```
