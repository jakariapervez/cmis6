```python
def findGLValueFromSection(section_data):
    rl_list=list(section_data['RL (m)'] )
    remark_list=list(section_data['Remarks'] )
    chainage_list=list(section_data['Chainage (m)'])
    lb_index=remark_list.index('LB')
    gl=rl_list[lb_index]
    chainage=chainage_list[lb_index]
    myvalue={"Chainage":chainage,"GL": gl}
    return myvalue
             
def findGL(input_df,output_df):
    sectionlist=list(output_df['SectionNo'])
    print(sectionlist)
    for section in sectionlist:
        x=input_df.loc[input_df['Section No']==section]
        myvalue=findGLValueFromSection(x)
        gl=myvalue["GL"]
        chn=myvalue["Chainage"]
        print(x)
        gl_index=sectionlist.index(section)
        output_df.iloc[gl_index,1]=chn
        output_df.iloc[gl_index,2]=gl
        print("GL={}".format(gl))
    return output_df
    
       
    
```
